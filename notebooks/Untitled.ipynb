```python
import docx2txt
from pathlib import Path
import re
```

```python
data_dir = Path.cwd().parent / 'data'
docx_dir = data_dir / 'docx'
```

```python
# pattern = '合同编号：【(nhyk-\d{4}-\d{2}-\d{4})】'
pattern = '合同编号：【(.*)】，'
for p in docx_dir.glob('*.docx'):
    try:
        text = docx2txt.process(p)
    except:
        print(f'Error: {p}')
    
    match = re.search(pattern, text)
    if match:
        print(f'{match.group(1)},{p.name}')
    else:
        print(f'no match {p}')
    
```
