```python
import re
import win32com.client as win32
from win32com.client import constants
from pathlib import Path, PureWindowsPath
from shutil import copyfile
```

```python
# 数据文件目录
data_dir = Path.cwd().parent / 'data'
# doc 文件目录
doc_dir = data_dir / 'doc';
# 转换成 docx 的目录
docx_dir = data_dir / 'docx'
# 转换失败目录
convert_fail_dir = data_dir / 'fail'
```

```python
fn = '2#201房屋租赁协议补充协议（含附件物业服务细则）【适用于产业园区事业部物业管理优化方案】.doc'
ppp = doc_dir / fn
print(str(ppp))
```

```python
def save_as_docx(path):
    # Opening MS Word
    word = win32.gencache.EnsureDispatch('Word.Application')
    doc = word.Documents.Open(str(path))
    doc.Activate ()

    # Rename path with .docx
#     new_file_abs = os.path.abspath(path)
#     new_file_abs = re.sub(r'\.\w+$', '.docx', new_file_abs)
    new_file_abs = docx_dir / f'{path.stem}.docx'

    # Save and Close
    word.ActiveDocument.SaveAs(
        str(new_file_abs), FileFormat=constants.wdFormatXMLDocument
    )
    doc.Close(False)
```

```python
for p in doc_dir.glob('*.doc'):
    try:
        save_as_docx(p)
    except:
        print(f'convert {p} fail.')

print('convert complete.')
```
