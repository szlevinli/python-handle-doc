# 转换 doc 文件为 docx 文件

仅适用于 `macOS`

## 准备

1. 安装 [LibreOffice](https://www.libreoffice.org/download/download/).
2. Make soffice command of Libreoffice work on Mac OS.

    ```bash
    sudo curl https://gist.githubusercontent.com/pankaj28843/3ad78df6290b5ba931c1/raw/soffice.sh > /usr/local/bin/soffice && sudo chmod +x /usr/local/bin/soffice
    ```

````python
from pathlib import Path
import subprocess
from shutil import copyfile
````

````python
# 数据文件目录
data_dir = Path.cwd().parent / 'data'
# doc 文件目录
doc_dir = data_dir / 'doc';
# 转换成 docx 的目录
docx_dir = data_dir / 'docx'
# 转换失败目录
convert_fail_dir = data_dir / 'fail'
````

````python
for p in doc_dir.glob('*.doc'):
    r = subprocess.call(['soffice', '--headless', '--convert-to', 'docx', '--outdir', str(docx_dir), p])
    if r != 0:
        copyfile(p, convert_fail_dir)
        print(f'convert {p.name} fail.')
print(f'convert {str(doc_dir)} to {str(docx_dir)} complete.')
````
